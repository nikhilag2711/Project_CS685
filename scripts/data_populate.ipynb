```python
station_df = final_df1.append(final_df2).reset_index(drop=True)
station_df['Date'] = pd.to_datetime(station_df['Date'], format='%d-%m-%Y')
station_df = station_df.sort_values(by=['City', 'Station', 'Date']).reset_index(drop=True)
station_df = station_df.replace(['None', 0], [np.NaN, np.NaN])
station_df
```

```python
city_df = station_df.groupby(by=['City', 'Date'], as_index=False).mean()
city_df = city_df.sort_values(by=['City', 'Date']).reset_index(drop=True)
city_df = city_df.drop(columns=['Eth-Benzene'])
city_df = city_df.rename(columns={'Ozone': 'O3'})
city_df
```

```python
city_df["AQI"] = city_df.apply(lambda x: aqic.AQI_calculate(x['PM2.5'], x['PM10'], x['SO2'], x['NOx'], x['NH3'], x['CO'], x['O3']), axis=1)
city_df
```

```python
city_df["AQI_Bucket"] = city_df.apply(lambda x: aqic.AQI_bucket_calculate(x['AQI']), axis=1)
city_df = city_df[['City', 'Date', 'PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'AQI', 'AQI_Bucket']]
city_df
```
